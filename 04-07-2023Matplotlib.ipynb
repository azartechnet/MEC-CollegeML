```python
import numpy as np
```

```python
t1=np.arange(5)
print(t1)
print(t1**2)
print(t1**3)
print(t1**4)
plt.plot(t1,'r--',t1**2,'bs',t1**3,'g^',t1**4,'ro')
plt.show()
```

```python
x=[1,2,3,4]
plt.plot(x,[elem for elem in x],label='linear')
plt.plot(x,[elem for elem in x],label='squar')
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.grid(True)
plt.legend()
plt.savefig('myimg.png')
plt.show()
```

```python
#LinearRegression
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
#1.Collecting the dataset

data=pd.read_csv("Datasets/BostonHousing.csv")
data.head()
```

```python
#2.DataWrangling
#input or independent variable
X=data.iloc[:,0:13]
X
```

```python
Y=data['medv']
Y
```

```python
#3.DataAnalysis
#Correlation
import seaborn as sns
```

```python
c1=data.corr()
#c1
sns.heatmap(c1)
plt.xticks(rotation=90)
plt.yticks(rotation=0)
plt.show()
```

```python
#4.Train the algorithm
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=.3,random_state=10)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
#Creating the model
lin_model=LinearRegression()
lin_model
```

```python
#fit the model with the input and ouput
lin_model.fit(X_train,Y_train)
```

```python
#5.Test algorithm
predicated_house_price=lin_model.predict(X_test)
plt.scatter(Y_test,predicated_house_price)
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.show()
```

```python
pd.DataFrame({"Predicatedvalue":predicated_house_price,"KnownO/P":Y_test})
```
